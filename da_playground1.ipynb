```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

open_power_data_path = 'opsd_germany_daily.csv'

# Read data from csv file
opsd_daily = pd.read_csv(open_power_data_path)

print(opsd_daily.shape)
print(opsd_daily.dtypes)
opsd_daily.head()
```

```python
# Set column Date to be the index
opsd_daily_indexed = opsd_daily.set_index('Date')

print(opsd_daily_indexed.shape)
print(opsd_daily_indexed.dtypes)
print('----')
print(opsd_daily_indexed.info())
```

```python
# Read csv file and set column Date as index

opsd_daily_new = pd.read_csv(open_power_data_path, index_col=0, parse_dates=True)

opsd_daily_new.info()
opsd_daily_new.isnull().sum()
```

```python
opsd_daily_new.head(3)
```

```python
# Add columns with year, month, and weekday name
opsd_daily_new['Year'] = opsd_daily_new.index.year
opsd_daily_new['Month'] = opsd_daily_new.index.month
opsd_daily_new['Weekday Name'] = opsd_daily_new.index.day_name()

# Display smapling of 5 rows
opsd_daily_new.sample(5)
```

```python
# Time-based indexing
opsd_daily_new.loc['2014-01-20':'2014-01-22']
```

```python
opsd_daily_new.loc['2014-02']
```

```python
# Plot consumption
col_name = 'Consumption'

sns.set_theme(rc={'figure.figsize': (11, 4)})
opsd_daily_new[col_name].plot(linewidth=0.5)
plt.show()
```

```python
cols_to_plot = ['Consumption', 'Solar', 'Wind']

axes = opsd_daily_new[cols_to_plot].plot(
    marker='.', alpha=0.5, linestyle='None',
    figsize=(11, 9), subplots=True
)

for ax in axes:
    ax.set_ylabel('Daily Total (GWh)')

plt.show()
```

```python
# Seasonality
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)

for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
    sns.boxplot(data=opsd_daily_new, x='Month', y=name, ax=ax)
    ax.set_ylabel('Gwh')
    ax.set_title(name)

    # Remove the automatic x_axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')

plt.show()
```

```python
# Create date_range from strings

date_range = pd.date_range('1998-03-10', '1998-03-15', freq='D')

print(date_range)
```

```python
# Implement forward fill

# Time sample 3 days
time_sample = pd.to_datetime(['2013-02-03', '2013-02-06', '2013-02-08'])

# Get consumption for time_sample
consum_sample = opsd_daily_new.loc[time_sample, ['Consumption']].copy()

print(consum_sample)
print(type(consum_sample))
```

```python
# Create frequency for consumption
consum_freq = consum_sample.asfreq('D')
print(consum_freq)
```

```python
# Forward fill 
consum_freq['Consumption - Forward Fill'] = consum_sample.asfreq('D', method='ffill')

consum_freq
```

```python
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']

opsd_weekly_mean = opsd_daily_new[data_columns].resample('W').mean()
opsd_weekly_mean.head()
```

```python
print(opsd_daily_new.shape[0])
print(opsd_weekly_mean.shape[0])
```

```python
# Start and end of the date range to extract
start, end = '2017-01', '2017-06'

# Plot daily and weekly resampled time series together
fig, ax = plt.subplots()
ax.plot(opsd_daily_new.loc[start:end, 'Solar'], marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'], marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (Gwh)')

ax.legend()
ax.set_title('Time Series for Solar for Daily and Weekly')
plt.show()
```

```python
# Compute the annual sums, setting the value to NaN for any where which has fewer than 360 days of data
opsd_annual = opsd_daily_new[data_columns].resample('A').sum(min_count=360)

opsd_annual
```

```python
# The default index of the resampled DataFrame is the last day of each year,
# To make life easier, set the index to the year component
opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
```

```python
opsd_annual.index.name = 'Year'
```

```python
opsd_annual
```

```python
# Compute the ratio of Wind+Solar to Consumption
opsd_annual['Wind+Solar/Consumption'] = opsd_annual['Wind+Solar'] / opsd_annual['Consumption']

opsd_annual.tail()
```

```python
# Plot from 2012 onwards, because there is no solar production data in earlier years

ax = opsd_annual.loc[2012:, 'Wind+Solar/Consumption'].plot.bar(color='C0')
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind + Solar Share of Annual Electricity Consumption')
plt.xticks(rotation=0)

plt.show()
```

```python
# Compute the centered 7-day rolling mean
opsd_7d = opsd_daily_indexed[data_columns].rolling(7, center=True).mean()

opsd_7d.head()
```

```python
data_columns
```

```python
import matplotlib.dates as mdates

# The min_periods=360 argument accounts for a few isolated missing days in the
# wind and solar production time series.

opsd_365d = opsd_daily_indexed[data_columns].rolling(window=365,
                                                     center=True,
                                                     min_periods=360).mean()

# Plot daily, 7-day rolling mean, and 365-day roling mean time series
fig, ax = plt.subplots()
ax.plot(opsd_daily_indexed['Consumption'], marker='.', markersize=2,
        color='0.6', linestyle='None', label='Daily')
ax.plot(opsd_7d['Consumption'], linewidth=2, label='7-d rolling mean')
ax.plot(opsd_365d['Consumption'], color='0.2', linewidth=3,
        label='Trend (365-d rolling mean)')

# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (Gwh)')
ax.set_title('Trends in Electricity Consumption')
plt.show()
```

```python
# Plot 365-day rolling mean time series of wind and solar power
fig, ax = plt.subplots()

for energy_type in ['Wind', 'Solar', 'Wind+Solar']:
    ax.plot(opsd_365d[energy_type], label=energy_type)
    # Set x-ticks to yearly interval, adjust y0axis limits, and legend and labels
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.set_ylim(0, 400)
    ax.legend()
    ax.set_ylabel('Production (Gwh)')
    ax.set_title('Trends in Electricity Production (365-d Rolling Means)')

plt.show()
```
